```python
from cloud.dependencies.clients import resolve_calendar_event_client

client = resolve_calendar_event_client()
```

```python
from datetime import date
from sebastian.domain.calendar import Calendars

test_title = "[DELETE TEST] delete_event showcase"
test_date = date(2026, 4, 15)

client.create_event(Calendars.Primary, test_title, test_date)
print("Event created")

events = client.get_events(Calendars.Primary, q=test_title, time_min=test_date)
print(f"Found {len(events)} event(s): {events[0].id}")

client.delete_event(Calendars.Primary, events[0].id)
print("Event deleted")

events_after = client.get_events(Calendars.Primary, q=test_title, time_min=test_date)
print(f"Events after deletion: {len(events_after)} (expected 0)")
```
