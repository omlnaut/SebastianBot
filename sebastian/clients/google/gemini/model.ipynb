# Gemini client

```python
from cloud.helper.secrets import SecretKeys, get_secret
from sebastian.clients.google.gemini.credentials import GeminiApiKey


key = get_secret(SecretKeys.GeminiApiKey)
```

```python
from google import genai
```

```python
genai.__version__
```

```python
client = genai.Client(api_key=key.api_key)
```

```python
for model in client.models.list():
    print(model.name)
    break
```

```python
from pydantic import BaseModel, Field


class Joke(BaseModel):
    genre: str = Field(description="The genre of the joke, e.g., 'programming', 'knock-knock', etc.")
    joke: str = Field(description="The content of the joke.")
```

```python
response = client.interactions.create(
                model="gemini-3.5-flash",
                input="tell me a joke about programmers",
                response_format={
                    "type": "text",
                    "mime_type": "application/json",
                    "schema": Joke.model_json_schema(),
                },
    )
```

```python
parsed = Joke.model_validate_json(response.output_text)
```
