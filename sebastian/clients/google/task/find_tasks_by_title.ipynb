```python
from cloud.dependencies.clients import resolve_google_task_client


client = resolve_google_task_client()
```

```python
s = client._service
```

```python
from sebastian.protocols.google_task.models import TaskListIds

# Collect all matching tasks
matching_tasks = []

# Get tasks from the Default list
tasks_response = s.tasks().list(tasklist=TaskListIds.Default.value, showCompleted=False).execute()
tasks = tasks_response.get('items', [])

# Filter tasks that start with "Paket abholen:"
for task in tasks:
    if task.get('title', '').startswith('Paket abholen:'):
        matching_tasks.append({
            'list_name': 'Default',
            'task_id': task['id'],
            'title': task['title'],
            'status': task.get('status'),
            'due': task.get('due')
        })

# Display results
for task in matching_tasks:
    print(f"List: {task['list_name']}")
    print(f"  Title: {task['title']}")
    print(f"  Status: {task['status']}")
    print(f"  Due: {task['due']}")
    print()

print(f"Found {len(matching_tasks)} matching tasks")
```

```python
# Mark the matching tasks as completed
for task in matching_tasks:
    updated_task = s.tasks().patch(
        tasklist=TaskListIds.Default.value,
        task=task['task_id'],
        body={'status': 'completed'}
    ).execute()
    print(f"Marked '{task['title']}' as completed.")
```
