```python
from googleapiclient.discovery import build

service = build("calendar", "v3", credentials=creds)
```

```python
import datetime


tomorrow = datetime.date.today() + datetime.timedelta(days=1)
start_time = datetime.datetime.combine(tomorrow, datetime.time(10, 0, 0)).isoformat()
end_time = datetime.datetime.combine(tomorrow, datetime.time(11, 0, 0)).isoformat()

event_body = {
    'summary': 'Python API Test Event',
    'location': 'Online',
    'description': 'This event was created automatically via the Google Calendar API.',
    'start': {
        'dateTime': start_time,
        'timeZone': 'America/Los_Angeles', # Change to your timezone
    },
    'end': {
        'dateTime': end_time,
        'timeZone': 'America/Los_Angeles',
    },
    'reminders': {
        'useDefault': False,
        'overrides': [
            {'method': 'email', 'minutes': 24 * 60},
            {'method': 'popup', 'minutes': 10},
        ],
    },
}
```

```python
event = service.events().insert(calendarId='primary', body=event_body).execute()
```
