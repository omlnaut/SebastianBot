# WinSim invoices

Goal:
- Once a day, check for new winsim invoice mails, download pdf attachment to ram
- upload to google drive into winsim folder
    - use yyyy-mm_winsim.pdf as filename

## Download to ram

```python
from cloud.dependencies.clients import resolve_gmail_client


gm = resolve_gmail_client()
```

```python
from datetime import datetime
from sebastian.clients.google.gmail.query_builder import GmailQueryBuilder


query = GmailQueryBuilder().from_email("no-reply@winsim.de").has_attachment("pdf").after_date(datetime(2025, 10, 1)).build()
```

```python
pdfs = gm.download_pdf_attachments(query)
```

```python
len(pdfs)
```

## Upload to drive

```python
from cloud.dependencies.clients import resolve_google_drive_client


drive = resolve_google_drive_client()
```

```python
winsim_dir_id = "1VGX5Wt8D3huZm3vVemjI3C6zz6W38PJr"
```

```python
from sebastian.clients.google.drive.models import UploadFileRequest


drive.upload_file(UploadFileRequest(content=pdfs[0].data, filename="2025-10_winsim.pdf", folder_id=winsim_dir_id))
```
