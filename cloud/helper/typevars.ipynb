```python
from abc import ABC, abstractmethod
from typing import Self, TypeVar

TRequest = TypeVar("TRequest")
TBase = TypeVar("TBase", bound="Base")

class Base[TRequest](ABC):
    @classmethod
    @abstractmethod
    def from_app(cls, request : TRequest) -> Self:
        """Create an instance of the class from an application-layer model."""
        ...

class Derived(Base[int]):
    @classmethod
    def from_app(cls, request: int) -> Self:
        print("Implemented in Derived: ", request)
        return cls()
```
